```python
import matplotlib.pyplot as plt
from retrieve.get_new_data import get_yield

df_1 = get_yield(term=1)
df_1 = df_1.ffill()
print(df_1.columns)

plt.figure(figsize=(12, 6))

for col in df_1.columns:
    if col == 'JAPAN':
        continue
    plt.plot(df_1.index, df_1[col], label=col)

plt.title("1-Year Yields")
plt.xlabel("Date")
plt.ylabel("Yield")
plt.legend()
plt.show()
```
